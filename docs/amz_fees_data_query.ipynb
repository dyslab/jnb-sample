Doc title: **Amazon Fees Data Query Snippet**

Article notes: All data came from 'Reports/Fulfillment/Fee Preview' @Amazon Seller Central. Certain columns with sensitive data have been removed from the exported csv file.

文章备注：本文使用的数据来源于亚马逊后台的“数据报告/库存和销售报告/费用预览”报表，删除了其中的部分敏感数据列。

Last modified date: 2019-07-19 17:02:01 

```python
# 引入pandas数据分析模块
import pandas as pd

# 读取数据源CSV文件
workdf = pd.read_csv('amz_fees_data_20190719.csv', index_col=0).sort_index()
```

```python
q_sku = input('请输入要查询的亚马逊商品SKU（如：0701, 0731-Z）：') # 注意：是商品SKU，而非产品型号。

def listInchToCM(inch_list):
    return [round(len * 2.54, 2) for len in inch_list]
    
def listPoundToKG(pound_list):
    return [round(pound * 0.4535924, 2) for pound in pound_list]

querydf = workdf.filter(like=q_sku.upper(), axis=0)

resultdf = pd.DataFrame({
    '商品SKU': querydf.index.tolist(),
    '规格类型': querydf['product-size-tier'].tolist(),
    '长': listInchToCM(querydf['longest-side'].tolist()),
    '宽': listInchToCM(querydf['median-side'].tolist()),
    '高': listInchToCM(querydf['shortest-side'].tolist()),
    '重量': listPoundToKG(querydf['item-package-weight'].tolist()),
    '售价': querydf['sales-price'].tolist(),
    '佣金': querydf['estimated-referral-fee-per-unit'].tolist(),
    'FBA配送费': querydf['expected-fulfillment-fee-per-unit'].tolist()
})
resultdf
```

查询结果备注项：

- 长度、重量为亚马逊实测计量数据。

- 长度单位为：CM

- 重量单位为：KG

- 货币单位为：美元
