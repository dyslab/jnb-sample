Article notes: Data source came from 'Reports/Fulfillment/Fee Preview' @Amazon Seller Central. Certain sensitive columns has been removed from the exported csv file.

文章备注：本文使用的数据来源于亚马逊后台的“数据报告/库存和销售报告/费用预览”报表，删除了其中的部分敏感数据列。

```python
# 引入pandas数据分析模块
import pandas as pd

# 读取数据源CSV文件
workdf = pd.read_csv('amz_fees_data_20190719.csv', index_col=0).sort_index()
```

```python
q_sku = input('请输入要查询的亚马逊商品SKU（如：0701, 0731-Z）：') # 注意：是商品SKU，而非产品型号。

def InchesToCM(inch_list):
    cm_list = []
    for len in inch_list:
        cm_list.append(len * 2.54)
    return cm_list

def PoundToKG(pound_list):
    kg_list = []
    for pound in pound_list:
        kg_list.append(pound * 0.4535924)
    return kg_list

querydf = workdf.filter(like=q_sku.upper(), axis=0)

resultdf = pd.DataFrame({
    '商品SKU': querydf.index.tolist(),
    '类型': querydf['product-size-tier'].tolist(),
    '长': InchesToCM(querydf['longest-side'].tolist()),
    '宽': InchesToCM(querydf['median-side'].tolist()),
    '高': InchesToCM(querydf['shortest-side'].tolist()),
    '重量': PoundToKG(querydf['item-package-weight'].tolist()),
    '售价': querydf['sales-price'].tolist(),
    '佣金': querydf['estimated-referral-fee-per-unit'].tolist(),
    'FBA配送费': querydf['expected-fulfillment-fee-per-unit'].tolist()
})
resultdf
```

查询结果备注项：

- 长度、重量为亚马逊实测计量数据。

- 长度单位为：CM

- 重量单位为：KG

- 货币单位为：美元
