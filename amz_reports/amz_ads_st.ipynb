```python
# 引入pandas数据分析模块
import pandas as pd

# 数据范例：美国站，月度数据
workdf = pd.read_excel('data/amz_ads_st_us_201908.xlsx', usecols=['广告活动名称', '广告组名称', '客户搜索词', '展现量', '点击量', '花费', '7天总销售额(￥)', '7天总销售量(#)'])
```

# 广告组数据排序（以销量为取值标准）

```python
# Manipulating data and column names.
impr10_df = workdf.groupby(['广告活动名称', '广告组名称']).sum().sort_values(by=['7天总销售量(#)'], ascending=False).head(20)
impr10_df['总销售额'] = impr10_df['7天总销售额(￥)']
impr10_df['总销售量'] = impr10_df['7天总销售量(#)']
impr10_df = impr10_df[['花费', '总销售额', '总销售量']]
impr10_df['卖出1件商品的平均花费'] = round(impr10_df['花费'] / impr10_df['总销售量'], 2)

impr10_df.sort_values(by=['卖出1件商品的平均花费'])
```

此表解决的问题为：

- 找出最具性价比（卖出1件商品花费越小，则性价比越高）的广告组。

# 客户搜索词数据排序（以展现量为取值标准）¶

```python
# Manipulating data and column names.
st10_df = workdf.groupby(['广告组名称', '客户搜索词']).sum().sort_values(by=['展现量'], ascending=False).head(20)
st10_df['总销售额'] = st10_df['7天总销售额(￥)']
st10_df['总销售量'] = st10_df['7天总销售量(#)']
st10_df = st10_df[['展现量', '点击量', '花费', '总销售额', '总销售量']]
st10_df['性价比(总销售额/花费)'] = round(st10_df['总销售额'] / st10_df['花费'], 2)

st10_df.sort_values(by=['性价比(总销售额/花费)'], ascending=False)
```

此表解决的问题为：

- 找出最具性价比的客户搜索词和广告组。
