```python
# 引入pandas数据分析模块
import pandas as pd

# 数据范例：美国站，月度数据
workdf = pd.read_excel('data/amz_ads_ap_us_201908.xlsx', usecols=['广告活动名称', '广告组名称', '广告SKU', '展现量', '点击量', '花费', '7天总销售额(￥)', '7天总销售量(#)', '7天内其他SKU销售量(#)', '7天内其他SKU销售额(￥)'])
```

# 广告组数据排序（以销量为取值标准）

*注：此部分结果请查看 [amz_ads_target.ipynb](amz_ads_target.ipynb) 中相关内容。*

此表解决的问题为：

- 找出展现量最高的广告组。

- 找出最具性价比的广告组。（卖出1件商品花费越小，则性价比越高）

# 广告产品数据排序（以展现量为取值标准）

## 最具性价比的广告产品以及其所在的广告组

```python
# Manipulating data and column names.
ap_df = workdf.groupby(['广告活动名称', '广告组名称', '广告SKU']).sum().sort_values(by=['展现量'], ascending=False).head(20)
ap_df['总销售额'] = ap_df['7天总销售额(￥)']
ap_df['总销售量'] = ap_df['7天总销售量(#)']
ap_df = ap_df[['展现量', '点击量', '花费', '总销售额', '总销售量']]
ap_df['性价比(总销售额/花费)'] = round(ap_df['总销售额'] / ap_df['花费'], 2)

ap_df.sort_values(by=['性价比(总销售额/花费)'], ascending=False)
```

此表解决的问题为：

- 找出最具性价比的广告产品以及其所在的广告组。

## 最具性价比的广告产品（不考虑其所在的广告组）

```python
# Manipulating data and column names.
ap_df = workdf.groupby(['广告SKU']).sum().sort_values(by=['展现量'], ascending=False).head(20)
ap_df['总销售额'] = ap_df['7天总销售额(￥)']
ap_df['总销售量'] = ap_df['7天总销售量(#)']
ap_df = ap_df[['展现量', '点击量', '花费', '总销售额', '总销售量']]
ap_df['性价比(总销售额/花费)'] = round(ap_df['总销售额'] / ap_df['花费'], 2)

ap_df.sort_values(by=['性价比(总销售额/花费)'], ascending=False)
```

此表解决的问题为：

- 找出最具性价比的广告产品（不考虑其所在的广告组）。
