```python
# 引入pandas数据分析模块
import pandas as pd

# 数据范例：美国站，2019年6月份数据
workdf = pd.read_csv('data/amz_fba_inventory_month_us_201906.csv')
```

# 各商品SKU期末库存数量

```python
target_df = workdf.groupby(['sku']).sum().sort_index()

target_df.head()
```

```python
print('期末库存型号数量：{1:.0f}种， 库存商品总数量：{0:.0f}件。'.format(workdf.sum()['end-quantity'], target_df.count()['end-quantity']))
```

```python
# Query
sku = input('请输入商品SKU（支持模糊查询）：')

target_df.filter(like=sku, axis=0)
```
