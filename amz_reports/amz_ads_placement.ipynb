```python
# 引入pandas数据分析模块
import pandas as pd

# 数据范例：美国站，月度数据
workdf = pd.read_excel('data/amz_ads_placement_us_201907.xlsx', usecols=['广告活动名称', '竞价策略', '放置', '展现量', '点击量', '花费', '7天总销售额(￥)', '7天总销售量(#)'])
```

# 广告位置数据排序（以展现量为取值标准）

## 点击率表现最好的广告位置

```python
# Manipulating data and column names.
placement_df = workdf.groupby(['广告活动名称', '放置']).sum().sort_values(by=['展现量'], ascending=False).head(20)
placement_df['总销售额'] = placement_df['7天总销售额(￥)']
placement_df['总销售量'] = placement_df['7天总销售量(#)']
placement_df = placement_df[['展现量', '点击量', '花费', '总销售额', '总销售量']]
placement_df['点击率(点击量/展现量)'] = round(placement_df['点击量'] / placement_df['展现量'] * 100, 2)

placement_df.sort_values(by=['点击率(点击量/展现量)'], ascending=False)
```

解决问题：

- 在具备一定展现量基础上，不同位置的点击率哪个表现更好。

## 不同位置的性价比（花费、销量、销售额）对比

```python
# Manipulating data and column names.
placement_df = workdf.groupby(['放置']).sum().sort_values(by=['展现量'], ascending=False).head(20)
placement_df['总销售额'] = placement_df['7天总销售额(￥)']
placement_df['总销售量'] = placement_df['7天总销售量(#)']
placement_df = placement_df[['展现量', '点击量', '花费', '总销售额', '总销售量']]
placement_df['性价比(总销售额/花费)'] = round(placement_df['总销售额'] / placement_df['花费'], 2)

placement_df.sort_values(by=['性价比(总销售额/花费)'], ascending=False)
```

此表解决的问题为：

- 找出最具性价比的广告位置。
